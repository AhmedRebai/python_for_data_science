# Comment télécharger des données à partir du site internet de google finance?

Ici, on va utiliser quelques modules en python pour faciliter réaliser cette tâches.

## Installation des modules:

Sous Linux (ubuntu), il suffit d'ouvrir un terminal ctrl + alt + t puis écrire les commandes suivantes: 

* Pour installer pandas
pip install pandas
* Pour installer pandas_datareader
pip install pandas_datareader
* Pour installer numpy
pip install numpy 
* Pour installer matplotlib
pip install matplotlib

```python
import pandas as pd
import datetime
import pandas_datareader.data as web

import matplotlib.pyplot as plt
from matplotlib import style

# Ici on choisit le style
# style.use('ggplot')
style.use('fivethirtyeight')


start = datetime.datetime(2013, 1, 1) 
end = datetime.datetime.now() 

df = web.DataReader("GOOGL", "iex", start, end) 

print(df.head()) 


df['open'].plot()
df['close'].plot()
df['high'].plot()
df['low'].plot()

plt.show()
```
